```python
import pandas as pd
import seaborn as srn
import statistics as sts
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('jogadores_lol.csv')
dataset.head(50), (49)
```

```python
dataset.head()
```

```python
agrupado = dataset.groupby(['Função']).size()
agrupado
```

```python
agrupado.plot.bar(color = 'skyblue')
plt.title('Funções mais escolhidas')
plt.xlabel('Funções')
plt.ylabel('Quantidade')
plt.xticks(rotation=22)
plt.grid(axis='y', linestyle='--', alpha=0.5)  
plt.tight_layout() 
plt.show()
```

```python
df['Função'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))
plt.tight_layout()
```

```python
df['Região'] = df['Região'].replace({
    'BR': 'Brasil',
    'KR': 'Coreia do Sul',
    'EUNE': 'Europa Oriental',
    'EUW' : 'Europa Ocidental',
})
```

```python
agrupado = df['Região'].value_counts()
agrupado
```

```python
agrupado.plot.bar(color = 'skyblue')
plt.title('Regiões mais populares')
plt.xlabel('Regiões')
plt.ylabel('Quantidade')
plt.xticks(rotation=22)
plt.grid(axis='y', linestyle='--', alpha=0.5)  
plt.tight_layout() 
plt.show()
```

```python


plt.figure(figsize=(10,6))
plt.scatter(df['Horas_Jogadas'], df['Taxa_Vitória'], color='green', alpha=0.6)
plt.title("Relação entre Horas Jogadas e Taxa de Vitória")
plt.xlabel("Horas Jogadas")
plt.ylabel("Taxa de Vitória")
plt.grid(True)
plt.show()
```

```python
correlacao = df['Horas_Jogadas'].corr(df['Taxa_Vitória'])
print(f"Correlação entre horas jogadas e taxa de vitória: {correlacao:.2f}")
```
